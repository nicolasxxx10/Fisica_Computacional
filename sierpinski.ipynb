```python
import random
import matplotlib.pyplot as plt

# ======================================
# FUNCION PARA ESCOGER UN VERTICE
# ======================================

def vertice_aleatorio(vertices):

    return random.choice(vertices)

# ======================================
# FUNCION PARA CALCULAR EL PUNTO MEDIO
# ======================================

def punto_medio(punto, vertice):

    x = (punto[0] + vertice[0]) / 2
    y = (punto[1] + vertice[1]) / 2

    return (x, y)

# ======================================
# VERTICES DEL TRIANGULO
# ======================================

A = (0, 0)
B = (1, 0)
C = (0.5, 0.866)

vertices = [A, B, C]

# ======================================
# PUNTO INICIAL
# ======================================

punto = (0.25, 0.3)

# ======================================
# LISTAS PARA GUARDAR PUNTOS
# ======================================

x_puntos = []
y_puntos = []

# ======================================
# NUMERO DE ITERACIONES
# ======================================

N = 50000

# ======================================
# GENERACION DEL FRACTAL
# ======================================

for i in range(N):

    vertice = vertice_aleatorio(vertices)

    punto = punto_medio(punto, vertice)

    x_puntos.append(punto[0])
    y_puntos.append(punto[1])

# ======================================
# GRAFICA
# ======================================

plt.figure(figsize=(7,7))

plt.scatter(x_puntos, y_puntos, s=0.1)

plt.title("Triangulo de Sierpinski")

plt.axis("equal")

plt.show()
```
