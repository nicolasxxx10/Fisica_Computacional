```python
import matplotlib.pyplot as plt

#=========================================================
# LISTAS
#=========================================================

h = []

exacta = []

cajas = []
trapecio = []
simpson = []

#=========================================================
# LEER ARCHIVO
#=========================================================

with open("datos_integracion.txt", "r") as archivo:

    next(archivo)

    for linea in archivo:

        datos = linea.split()

        h.append(float(datos[0]))

        exacta.append(float(datos[1]))

        cajas.append(float(datos[2]))
        trapecio.append(float(datos[3]))
        simpson.append(float(datos[4]))

#=========================================================
# GRAFICA
#=========================================================

plt.figure(figsize=(10,6))

plt.plot(h, exacta,
         marker='o',
         color='red',
         linewidth=3,
         label='Exacta')

plt.plot(h, cajas,
         marker='s',
         color='blue',
         linestyle='--',
         label='Cajas')

plt.plot(h, trapecio,
         marker='^',
         color='orange',
         linestyle='--',
         label='Trapecio')

plt.plot(h, simpson,
         marker='d',
         color='green',
         linestyle='--',
         label='Simpson')

#=========================================================
# CONFIGURACION
#=========================================================

plt.xlabel("h")

plt.ylabel("Integral")

plt.title("Comparacion de Metodos de Integracion")

plt.grid(True)

plt.legend()


#=========================================================
# GUARDAR
#=========================================================

plt.savefig("Integrales.png",
            dpi=300)

plt.show()
```
