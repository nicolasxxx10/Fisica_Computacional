```python
import matplotlib.pyplot as plt

#=========================================================
# LISTAS
#=========================================================

h = []

error_cajas = []

error_trapecio = []

error_simpson = []

#=========================================================
# LEER ARCHIVO
#=========================================================

with open("datos_integracion.txt", "r") as archivo:

    next(archivo)

    for linea in archivo:

        datos = linea.split()

        h.append(float(datos[0]))

        error_cajas.append(abs(float(datos[5])))

        error_trapecio.append(abs(float(datos[6])))

        error_simpson.append(abs(float(datos[7])))

#=========================================================
# ERRORES MAXIMOS
#=========================================================

error_max_cajas = max(error_cajas)

error_max_trapecio = max(error_trapecio)

error_max_simpson = max(error_simpson)

print("Error maximo cajas =", error_max_cajas)

print("Error maximo trapecio =", error_max_trapecio)

print("Error maximo simpson =", error_max_simpson)

#=========================================================
# GRAFICA
#=========================================================

plt.figure(figsize=(10,6))

plt.plot(h, error_cajas,
         marker='s',
         color='blue',
         linestyle='--',
         label='Cajas')

plt.plot(h, error_trapecio,
         marker='^',
         color='orange',
         linestyle='--',
         label='Trapecio')

plt.plot(h, error_simpson,
         marker='d',
         color='green',
         linestyle='--',
         label='Simpson')

#=========================================================
# CONFIGURACION
#=========================================================

plt.xlabel("h")

plt.ylabel("Error Maximo")

plt.title("Error de los Metodos Numericos")

plt.grid(True)

plt.legend()



#=========================================================
# GUARDAR
#=========================================================

plt.savefig("Errores.png",
            dpi=300)

plt.show()
```
