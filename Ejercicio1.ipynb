```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([95, 110, 90])

coeficientes = np.polyfit(x, y, 2)
A, B, C = coeficientes

x_max = -B / (2 * A)
y_max = A * x_max**2 + B * x_max + C

print(f"El coeficiente A es: {A:.2f}")
print(f"La posición x del máximo real es: {x_max:.2f}")
print(f"El valor máximo de la parábola es: {y_max:.2f}")
```
