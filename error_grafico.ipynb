```python
import matplotlib.pyplot as plt

#Listas vacias para almacenar los datos
x = []
error = []

#Abrir archivo
with open("error.txt") as archivo:

    #Leer cada linea del archivo
    for line in archivo:
        
        #Separar los valores de cada linea
        valores = line.split()

        #Agregar los valores a las listas correspondientes
        x.append(float(valores[0]))
        error.append(float(valores[1]))

#Graficar
plt.plot(x, error, color="black", linestyle="-")
plt.xlabel("x")
plt.ylabel("Error")
plt.grid()
plt.savefig("Error")
plt.show()
```
