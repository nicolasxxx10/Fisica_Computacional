```python
import numpy as np
import matplotlib.pyplot as plt

#  Leer datos de R_rms
datos = np.loadtxt("datos.txt")

N = datos[:,0]
R = datos[:,1]

#  Grafica 1: R_rms vs N
plt.figure()
plt.plot(N, R, 'o-', label="Simulacion")
plt.plot(N, np.sqrt(N), '--', label="√N")

plt.xlabel("N")
plt.ylabel("R_rms")
plt.title("Caminata aleatoria: R_rms vs N")
plt.savefig('Caminata.png')
plt.legend()
plt.grid()

#  Leer trayectoria
tray = np.loadtxt("trayectoria.txt")

x = tray[:,0]
y = tray[:,1]

#  Grafica 2: trayectoria
plt.figure()
plt.plot(x, y)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Trayectoria de la caminata aleatoria")
plt.savefig('Tray.Caminata.png')

plt.grid()

plt.show()
```
