```python
import matplotlib.pyplot as plt

#Listas vacias para almacenar los datos
h = []
error_max = []

#Abrir archivo
with open("errores_h.txt") as archivo:

    #Leer cada linea del archivo
    for line in archivo:
        
        #Separar los valores de cada linea
        valores = line.split()

        #Agregar los valores a las listas correspondientes
        h.append(float(valores[0]))
        error_max.append(float(valores[1]))

#Graficar
plt.plot(h, error_max, color="black", linestyle="-")
plt.xlabel("h")
plt.ylabel("Error Máximo")
plt.grid()
plt.savefig("Error_h.png")
plt.show()
```
