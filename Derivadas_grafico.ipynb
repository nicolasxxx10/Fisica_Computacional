```python
import matplotlib.pyplot as plt

#Listas vacias para almacenar los datos
x = []
derivada_ana = []
derivada_num = []

#Abrir archivo
with open("derivada.txt", "r") as archivo:

    #Leer cada linea del archivo
    for line in archivo:
        
        #Separar los valores de cada linea
        valores = line.split()

        #Agregar los valores a las listas correspondientes
        x.append(float(valores[0]))
        derivada_ana.append(float(valores[1]))
        derivada_num.append(float(valores[2]))

#Graficar
plt.plot(x,derivada_ana, color="blue",linestyle="--", label="Derivada Analitica")
plt.plot(x,derivada_num, color="red",linestyle="-.", label="Derivada Numerica")
plt.xlabel("x")
plt.ylabel("Derivada")
plt.grid()
plt.legend()
plt.savefig("Derivadas")        
plt.show()
```
